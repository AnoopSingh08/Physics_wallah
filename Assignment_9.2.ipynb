### Question2

```python
#A scatter plot is a type of data visualization that shows individual data points as dots on a two-dimensional plane. It is particularly useful for understanding the relationship between two continuous variables and identifying any patterns or trends in the data.

#In the given code, we are generating random data for two variables, x and y, using NumPy's random.normal function. The np.random.seed(3) line sets the random seed, ensuring that the generated data remains consistent each time the code is executed.

#Let's create a scatter plot using the generated x and y data and add title, xlabel, and ylabel to the plot using Matplotlib's Pyplot module:
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(3)

# Generate data for x and y
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y)

# Add title, xlabel, and ylabel to the plot
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()

#Explanation of the code:

#    The numpy module is imported as np.
#    The random seed is set to 3 using np.random.seed(3) for reproducibility.
#    The x and y data are generated using np.random.normal(0, 2, 50) and np.random.normal(0, 2, len(x)), respectively.
#    A scatter plot is created using plt.scatter(x, y).
#    The title is set using plt.title('Scatter Plot').
#    The x-axis label is set using plt.xlabel('X-axis').
#    The y-axis label is set using plt.ylabel('Y-axis').
#    Finally, the plot is displayed using plt.show().    
```

### Question3

```python
#The subplot() function in Matplotlib is used to create multiple plots within the same figure. It allows you to organize different subplots in a grid-like layout, making it easier to visualize and compare different datasets or aspects of your data. The function takes three arguments: num_rows, num_cols, and plot_number. These arguments determine the number of rows and columns in the subplot grid and the specific position of the current plot within that grid.

#Now, let's draw four line plots using the subplot() function with the given data:

import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 subplot grid and plot each line in a separate subplot
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line 4')

# Adjust the layout to prevent overlapping titles
plt.tight_layout()

# Display the plots
plt.show()
#In this code, we create a 2x2 subplot grid (2 rows and 2 columns). Each line is plotted in a separate subplot, and we set the title for each subplot using plt.title(). The plt.tight_layout() function is used to adjust the spacing between subplots, ensuring that the titles do not overlap with the neighboring subplots.

#When you run this code, it will display four line plots in a 2x2 grid, each representing a different dataset.
```

### Question4

```python
#A bar plot is a type of data visualization that represents categorical data with rectangular bars. Each bar's length or height is proportional to the values they represent, and the bars are usually aligned along the x-axis (vertical bar plot) or y-axis (horizontal bar plot). Bar plots are commonly used to compare different categories and visualize their quantities, making it easy to identify trends or disparities among the categories.

#The bar plot is used for the following purposes:

#    Comparison: It allows for easy comparison of the values between different categories.
#    Ranking: It can be used to rank categories based on their respective values.
#    Distribution: It helps visualize the distribution of categorical data.
#    Highlighting patterns: Bar plots can help identify patterns and outliers in the data.

#Let's plot a bar plot and a horizontal bar plot using the provided data:
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.figure(figsize=(8, 5))
plt.bar(company, profit, color='b')
plt.xlabel('Company')
plt.ylabel('Profit (Millions)')
plt.title('Profit by Company (Bar Plot)')
plt.grid(True, axis='y')
plt.show()

# Horizontal bar plot
plt.figure(figsize=(8, 5))
plt.barh(company, profit, color='g')
plt.ylabel('Company')
plt.xlabel('Profit (Millions)')
plt.title('Profit by Company (Horizontal Bar Plot)')
plt.grid(True, axis='x')
plt.show()
#Explanation of the code:

#    We import the necessary libraries, numpy as np, and matplotlib.pyplot as plt.
#    The data for companies and their corresponding profits are stored in arrays company and profit, respectively.
#    The first plot is created using plt.bar(company, profit, color='b'), which generates a vertical bar plot.
#    The second plot is created using plt.barh(company, profit, color='g'), which generates a horizontal bar plot.
#    We add labels and titles to the plots using plt.xlabel(), plt.ylabel(), and plt.title().
#    Grid lines are added using plt.grid(True, axis='y') for the vertical bar plot and plt.grid(True, axis='x') for the horizontal bar plot.
#    Finally, both plots are displayed using plt.show().
```

```python
# A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It shows a summary of the data's central tendency (median), spread, and potential outliers. The plot consists of a rectangular "box" that spans the interquartile range (IQR) of the data, with a line inside representing the median. Extending from the box are "whiskers," which typically cover a certain range from the minimum and maximum values. Data points that fall outside the whiskers are plotted individually as potential outliers.

# Box plots are commonly used for the following purposes:

#    Visualizing distribution: They provide a concise and clear way to visualize the spread of data and identify skewness or symmetry in the distribution.
#    Comparing datasets: Box plots allow for easy comparison of multiple datasets or groups, making it ideal for analyzing the variation between different categories.
#    Detecting outliers: The presence of outliers can be quickly identified in a box plot, making it helpful for data cleaning and anomaly detection.

#Let's plot a box plot for the provided data box1 and box2:

import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data for plotting
data = [box1, box2]

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot of Box 1 and Box 2')
plt.grid(True, axis='y')
plt.show()

# Explanation of the code:

#    We import the necessary libraries, numpy as np, and matplotlib.pyplot as plt.
#    The data box1 and box2 are generated using numpy's random.normal function, representing two datasets with different means and spreads.
#    The data is combined into a list data to plot them side by side in a single box plot.
#    The box plot is created using plt.boxplot(data, labels=['Box 1', 'Box 2']), where we provide labels for the two boxes.
#    We add labels and titles to the plot using plt.xlabel(), plt.ylabel(), and plt.title().
#    Grid lines are added using plt.grid(True, axis='y').
#    Finally, the box plot is displayed using plt.show().

# When you run this code, it will generate a box plot comparing the distribution of box1 and box2, helping you visualize their central tendency, spread, and potential outliers.
```
